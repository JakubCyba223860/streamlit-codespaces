```python
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

df = pd.read_csv('../Datasets/cleaned/L_P_B_M_C.csv')
```

```python
from sklearn.model_selection import train_test_split

X = pd.concat([df['Benches'], df['POI'], df['Lights'], df['Migrated']], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, df['CrimeCount'], test_size=0.2, random_state=42)


# Define the model
model = Sequential()

# Add layers to the model
model.add(Dense(32, activation='relu', input_shape=(X.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))  # No activation function for regression

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32)

# Evaluate the model
loss = model.evaluate(X_test, y_test)
print(f"Loss: {loss}")
```
